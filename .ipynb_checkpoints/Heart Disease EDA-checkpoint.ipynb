```python
sns.displot(df['Age'],kde=True)
```

## Pie chart for categorical data

```python
df.groupby('Sex').size().plot(kind='pie',autopct='%.0f')
```

```python
df.groupby('ChestPainType').size().plot(kind='pie',autopct='%.0f')
```

## Violin Plot

```python
sns.violinplot(df['Age'])
```

```python
sns.violinplot(y=df['Sex'],x=df['HeartDisease'])
```

```python
df.corr(numeric_only=True)
```

```python
sns.heatmap(df.corr(numeric_only=True) )
```

## Time for a Joint Plot

```python
sns.jointplot(x='Age',y='MaxHR',data=df)
```

```python
sns.jointplot(x='Age',y='Age',data=df,kind='hex')
```

```python
sns.jointplot(x='HeartDisease',y='Oldpeak',data=df)
```
